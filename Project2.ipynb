```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("../datasets/DC_Properties.csv", encoding = "ISO-8859-1") #1st column needs encoding to read pound/hashtag(#) symbol
df.head()
```

```python
#rename the pound/hashtag(#) column so it doesn't cause further issues
df.rename(columns={'#': 'Number'}, inplace =True)
```

```python
#validate
df.head()
```

```python
#set indices as  numbers
df.set_index('AYB', inplace=True)
df.head() 
```

```python
df.columns
```

```python
#make a scatterplot
sns.lmplot(x='BEDRM', y= 'PRICE', data=df)
```

```python
#add more details to scatterplot
sns.lmplot(x='BEDRM', y= 'PRICE', data=df, 
           fit_reg=False, #remove regression line
           hue='PRICE')   #color by evolution stage
```

```python
#make a scatterplot
sns.lmplot(x='CITY', y= 'PRICE', data=df)
```
